```python
df_country = pd.read_excel("Datasets/Zomatodataset/Country-Code.xlsx")
df_country.head()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
## Pie Chart - Top 3 countries that uses Zomato

plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%')
```

Observation:

Zomato maximum transactions are from 
1. India 
2. USA 
3. UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Now we want to find out how these are related.

Observation

If rating is between,
1. 4.5-4.9 ----> Excellent
2. 4.0-4.4 ----> Very Good
3. 3.5-3.9 ----> Good
4. 2.5-3.5 ----> Average
5. 1.8-2.4 ----> Poor
6. 0.0     ----> No rating

```python
ratings.head()
```

```python
# Barplot

sns.barplot(
    x="Aggregate rating", y="Rating Count", data=ratings,
    hue='Rating color',
    palette=['blue', 'red', 'orange', 'yellow', 'green', 'black']
)
```

Observation

1. Not Rated count is very high
2. Maximum number of rating are between 2.5 to 3.4
3. If some ratings are missing, use values between 2.5 to 3.4 as average and get it.

```python
# Count Plot

sns.countplot(
    x="Rating color", data=ratings,
    hue='Rating color',
    palette=['blue', 'red', 'orange', 'yellow', 'green', 'black']
)
```

```python
# The Countries that have given Zero ratings

final_df[ final_df['Aggregate rating'] == 0.0 ].groupby(['Country']).size().reset_index().rename(columns={0:"Ratings"})
```

```python
# Find out which currency is used by which country

final_df[ ['Country', 'Currency'] ].groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Which countries have Online Delivery option?

final_df[ final_df['Has Online delivery'] == 'Yes' ].Country.value_counts()
```

Observations:

1. Online Deliveries available in India and UAE

```python
# Create a Pie Chart for cities distribution

city_labels = final_df.City.value_counts().index
values = final_df.City.value_counts().values

plt.pie(values[:5], labels=city_labels[:5], autopct='%1.2f%%')
```

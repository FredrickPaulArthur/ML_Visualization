## Histograms

plotted using matplotlib's plt

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
import pandas as pd
data = load_breast_cancer() 

print(data.keys())
pd.DataFrame(data.data).sample(5, random_state=2049)
```

```python
plt.figure(figsize=(7,3))
plt.hist(data.data[:,0], bins=20)   # bins is for grouping of the datapoints in the graph
plt.xlabel("Mean Radius")
plt.ylabel("Frequency")
```

Looks like the data is roughly normally distributed

Now looking at the Frequency of the mean radius in malignant and benign samples

```python
plt.figure(figsize=(7,3))
plt.hist(data.data[data.target==0,0], bins=20, alpha=0.5, label="Malignant")
plt.hist(data.data[data.target==1,0], bins=20, alpha=0.5, label="Benign")
# condition specifies to extract 1st feature of the target column
plt.xlabel("Mean Radius")
plt.ylabel("Frequency")
plt.legend()

plt.show()
```

So the radius of Malignant tumors is large compared to benign ones

## Density plots

```python
import seaborn as sns
```

```python
plt.figure(figsize=(7,3))
sns.kdeplot(data.data[:,0], fill=True)
plt.xlabel("Mean Radius")
plt.ylabel("Density")
plt.show()
```

```python
plt.figure(figsize=(7,3))
sns.kdeplot(data.data[data.target==0,0], fill=True, label='Malignant')
sns.kdeplot(data.data[data.target==1,0], fill=True, label='Benign')
plt.xlabel("Mean Radius")
plt.ylabel("Density")
plt.show()
```

## Correlation Matrix (or HeatMap)

Correlation coefficients are linear relationships between two variables, here its 2 features in the dataset.

1  -  positive correlation

-1 -  negative correlation

0  -  no correlation

```python
import numpy as np

df = pd.DataFrame(data.data, columns=data.feature_names)
target = data.target

plt.figure(figsize=(7.5, 3.5))

corr = df.corr()
sns.set_theme(style='white')
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(
    corr,
    mask=mask, 
    cmap=cmap, 
    vmax=.3, 
    center=0,
    square=True, 
    linewidths=.5, 
    cbar_kws={
        "shrink": .5
    }
    
)
plt.show()
```

## Scatter Plot (or PairPlot)

```python
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
target = iris.target

plt.figure(figsize=(7.5, 3.5))
corr = df.corr()
sns.set_theme(style='white')
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(
    corr,
    mask=mask, 
    cmap=cmap, 
    vmax=.3, 
    center=0,
    square=True, 
    linewidths=.5, 
    cbar_kws={
        "shrink": .5
    }
    
)
plt.show()
```

```python
# create scatter matrix plot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')

# show plot
plt.show()
```
